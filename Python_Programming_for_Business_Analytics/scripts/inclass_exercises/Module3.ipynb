## Selecting Rows by Label: loc

`loc` = **loc**ation by label

````python
# By index label (default index is 0, 1, 2...)
df.loc[0]        # Row with index label 0
df.loc[0:4]      # Rows 0, 1, 2, 3, 4 (inclusive!)

# Rows and columns by name
df.loc[0:4, 'Name']              # Single column
df.loc[0:4, ['Name', 'Age']]     # Multiple columns
df.loc[0:4, 'Name':'Age']        # Column range
````

> ⚠️ **Warning:** `loc` includes the endpoint! `loc[0:4]` gives 5 rows.
`iloc` excludes it like normal Python! `iloc[0:4]` gives 4 rows.

## Single Condition Examples

````python
# Numeric comparisons
adults = df[df['Age'] >= 18]
children = df[df['Age'] < 18]
exact_age = df[df['Age'] == 30]

# Categorical comparisons
survivors = df[df['Survived'] == 1]
first_class = df[df['Pclass'] == 1]
women = df[df['Sex'] == 'female']

# Check results
print(f"Adults: {len(adults)}")
print(f"Survivors: {len(survivors)}")
print(f"Women: {len(women)}")
````

## Comparison Operators

| Operator | Meaning | Example |
|----------|---------|---------|
| `==` | Equal to | `df['Sex'] == 'female'` |
| `!=` | Not equal to | `df['Pclass'] != 3` |
| `<` | Less than | `df['Age'] < 18` |
| `>` | Greater than | `df['Fare'] > 100` |
| `<=` | Less than or equal | `df['Age'] <= 12` |
| `>=` | Greater than or equal | `df['Age'] >= 65` |

## Multiple Conditions: AND

Use `&` (ampersand) for AND:

````python
# Women who survived
women_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print(f"Women who survived: {len(women_survived)}")

# Adults in first class
adult_first = df[(df['Age'] >= 18) & (df['Pclass'] == 1)]
print(f"Adults in first class: {len(adult_first)}")

# Children who survived
child_survivors = df[(df['Age'] < 18) & (df['Survived'] == 1)]
print(f"Children who survived: {len(child_survivors)}")
````

> ⚠️ **Warning:** **Important:** Use `&` not `and`, and ALWAYS use parentheses around each condition!

## Multiple Conditions: OR

Use `|` (pipe) for OR:

````python
# First OR second class
upper_class = df[(df['Pclass'] == 1) | (df['Pclass'] == 2)]
print(f"Upper class: {len(upper_class)}")

# Very young or very old
extreme_ages = df[(df['Age'] < 5) | (df['Age'] > 70)]
print(f"Very young or old: {len(extreme_ages)}")
````

## Negation: NOT

Use `~` (tilde) for NOT:

````python
# NOT first class
not_first = df[~(df['Pclass'] == 1)]
print(f"Not first class: {len(not_first)}")

# NOT survived (same as Survived == 0)
died = df[~(df['Survived'] == 1)]
print(f"Did not survive: {len(died)}")
````

## Complex Conditions

````python
# Adult women in first class who survived
complex_filter = df[
    (df['Age'] >= 18) & 
    (df['Sex'] == 'female') & 
    (df['Pclass'] == 1) & 
    (df['Survived'] == 1)
]
print(f"Adult women in 1st class who survived: {len(complex_filter)}")

# Men OR children who survived
men_or_children_survived = df[
    ((df['Sex'] == 'male') | (df['Age'] < 18)) & 
    (df['Survived'] == 1)
]
````

## Filtering with isin()

Check if value is in a list:

````python
# Passengers in 1st or 2nd class
upper_class = df[df['Pclass'].isin([1, 2])]
print(f"1st or 2nd class: {len(upper_class)}")

# Embarked at Cherbourg or Southampton
ports = df[df['Embarked'].isin(['C', 'S'])]
print(f"Embarked at C or S: {len(ports)}")

# Specific ages
specific_ages = df[df['Age'].isin([25, 30, 35])]
````

## Filtering with between()

Check if value is in a range:

````python
# Age between 20 and 30 (inclusive)
young_adults = df[df['Age'].between(20, 30)]
print(f"Age 20-30: {len(young_adults)}")

# Fare between 10 and 50
mid_fare = df[df['Fare'].between(10, 50)]
print(f"Fare $10-$50: {len(mid_fare)}")
````

## Why String Operations Matter

Real-world text data is messy:
- Inconsistent capitalization: "Male", "MALE", "male"
- Extra whitespace: "  John  "
- Need to extract parts: "Mr. John Smith" → "Mr."
- Need to search: Find names containing "Mrs"

Pandas provides `.str` accessor for string operations.

````python
import pandas as pd

# 1) From a list of names (most common)
df = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie"]})

# 2) Empty DataFrame with a 'Name' column
df_empty = pd.DataFrame(columns=["Name"])

# 3) From a list of dicts
df2 = pd.DataFrame([{"Name": "Alice"}, {"Name": "Bob"}])

print(df)
````

## The .str Accessor

````python
# Access string methods on a column
# df['Name'].str.method_name()

# Examples:
df['Name'].str.lower()      # All lowercase
df['Name'].str.upper()      # All uppercase
df['Name'].str.len()        # Length of each string
df['Name'].str.strip()      # Remove whitespace
````

## Case Conversion

````python
# Create sample data
names = pd.Series(['  alice  ', 'BOB', 'Charlie', 'diana smith'])

print("Original:")
print(names.tolist())

print("\nLowercase:")
print(names.str.lower().tolist())

print("\nUppercase:")
print(names.str.upper().tolist())

print("\nTitle Case:")
print(names.str.title().tolist())

print("\nCapitalize (first letter only):")
print(names.str.capitalize().tolist())
````

## Whitespace Handling

````python
# Sample data with whitespace problems
dirty = pd.Series(['  Alice  ', 'Bob\t', '\nCharlie', '  Diana  '])

# Strip whitespace from both ends
clean = dirty.str.strip()
print(clean.tolist())  # ['Alice', 'Bob', 'Charlie', 'Diana']

# Strip only left side
dirty.str.lstrip()

# Strip only right side
dirty.str.rstrip()
````

## Searching Text: contains()

````python
# Find names containing "Mrs"
has_mrs = df['Name'].str.contains('Mrs')
mrs_passengers = df[has_mrs]
print(f"Mrs passengers: {len(mrs_passengers)}")

# Case-insensitive search
has_mrs = df['Name'].str.contains('mrs', case=False)

# Contains any of multiple patterns
has_title = df['Name'].str.contains('Mr|Mrs|Miss|Master')
````

## More Search Methods

````python
# Starts with
mr_names = df[df['Name'].str.startswith('Mr.')]

# Ends with
# Names ending with specific suffix
ends_son = df[df['Name'].str.endswith('son')]

# Match exact pattern (regex)
# df['Name'].str.match(pattern)
````

## Extracting Parts: split()

````python
# Split names by comma
# "Braund, Mr. Owen Harris" → ["Braund", " Mr. Owen Harris"]
parts = df['Name'].str.split(',')
print(parts.head())

# Get just the last name (first part after split)
df['LastName'] = df['Name'].str.split(',').str[0]
print(df['LastName'].head())

# Expand into multiple columns
name_parts = df['Name'].str.split(',', expand=True)
name_parts.columns = ['LastName', 'FirstAndTitle']
````

## Extracting with Regular Expressions

````python
# Extract title (Mr., Mrs., Miss, etc.)
df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.')

print(df['Title'].value_counts())
````

```
Mr          517
Miss        182

```

## Replacing Text

````python
# Simple replacement
df['Sex_short'] = df['Sex'].replace({'male': 'M', 'female': 'F'})

# String replacement with .str.replace()
# Remove all periods from names
df['Name_clean'] = df['Name'].str.replace('.', '', regex=False)

# Replace with regex
df['Name_clean'] = df['Name'].str.replace(r'[^\w\s]', '', regex=True)
````

## String Length and Counting

````python
# Length of each string
df['Name_length'] = df['Name'].str.len()
print(df['Name_length'].describe())

# Count occurrences of a character
df['comma_count'] = df['Name'].str.count(',')

# Count words (split by space)
df['word_count'] = df['Name'].str.split().str.len()
````

## Practical String Cleaning Example

````python
# Messy data
messy = pd.DataFrame({
    'product': ['  APPLE  ', 'banana', 'CHERRY', '  Orange '],
    'category': ['fruit', 'FRUIT', 'Fruit', 'fruit']
})

# Clean it
messy['product'] = messy['product'].str.strip().str.title()
messy['category'] = messy['category'].str.lower()

print(messy)
````

## Creating Date Data

## Date Arithmetic

## Calculating Age from Birthdate

````python
# Sample birthdate data
people = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'birthdate': pd.to_datetime(['1990-05-15', '1985-12-20', '2000-03-10'])
})

# Calculate age
today = pd.Timestamp.today()
people['age'] = (today - people['birthdate']).dt.days // 365

# More accurate age calculation
people['age_accurate'] = (
    today.year - people['birthdate'].dt.year - 
    ((today.month < people['birthdate'].dt.month) | 
     ((today.month == people['birthdate'].dt.month) & 
      (today.day < people['birthdate'].dt.day)))
)
````

# 📚 Part 7: GroupBy Operations

Split-Apply-Combine (30 minutes)

## What is GroupBy?

GroupBy follows the **split-apply-combine** pattern:

1. **Split**: Divide data into groups
2. **Apply**: Calculate something for each group
3. **Combine**: Put results back together

## Basic GroupBy Syntax

````python
# Average fare by passenger class
avg_fare = df.groupby('Pclass')['Fare'].mean()
print(avg_fare)
````

```
Pclass
1    84.154687
2    20.662183
3    13.675550
Name: Fare, dtype: float64
```

## GroupBy Structure

````python
# Syntax:
df.groupby('column_to_group_by')['column_to_aggregate'].function()

# Examples:
df.groupby('Sex')['Survived'].mean()      # Survival rate by sex
df.groupby('Pclass')['Age'].mean()        # Average age by class
df.groupby('Embarked')['Fare'].sum()      # Total fare by port
````

## Common Aggregation Functions

| Function | Description |
|----------|-------------|
| `mean()` | Average |
| `sum()` | Total |
| `count()` | Number of values |
| `min()` | Minimum |
| `max()` | Maximum |
| `std()` | Standard deviation |
| `median()` | Middle value |
| `first()` | First value |
| `last()` | Last value |
| `nunique()` | Number of unique values |

## Multiple Aggregations with agg()

````python
# Multiple functions on one column
df.groupby('Pclass')['Fare'].agg(['mean', 'min', 'max', 'count'])

# Different functions for different columns
summary = df.groupby('Pclass').agg({
    'Survived': 'mean',           # Survival rate
    'Age': ['mean', 'min', 'max'], # Age statistics
    'Fare': 'mean',               # Average fare
    'PassengerId': 'count'        # Count
})
print(summary)
````
