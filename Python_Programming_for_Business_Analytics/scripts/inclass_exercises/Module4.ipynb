## Conditionals Review

## Exercise 1.1: FizzBuzz Review

```python
# Print numbers 1-20
# Multiples of 3: "Fizz"
# Multiples of 5: "Buzz"
# Multiples of both: "FizzBuzz"

for num in range(1, 21):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

# 📚 Part 2: Pandas Review

Key Operations (30 minutes)

## Loading and Exploring

```python
import pandas as pd

# Load data
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Explore
print(df.shape)        # (891, 12)
print(df.columns)      # Column names
print(df.head())       # First 5 rows
print(df.info())       # Types + nulls
print(df.describe())   # Statistics
```

## Selecting Data

```python
# Single column (Series)
ages = df['Age']

# Multiple columns (DataFrame)
subset = df[['Name', 'Age', 'Survived']]

# Rows by index
first_row = df.iloc[0]
first_five = df.iloc[0:5]
```

## Filtering Data

```python
# Single condition
adults = df[df['Age'] >= 18]
survivors = df[df['Survived'] == 1]
women = df[df['Sex'] == 'female']

# Multiple conditions (use & and |)
adult_women = df[(df['Age'] >= 18) & (df['Sex'] == 'female')]
first_or_second = df[(df['Pclass'] == 1) | (df['Pclass'] == 2)]
```

## GroupBy Operations

```python
# Average by group
df.groupby('Pclass')['Fare'].mean()

# Survival rate by sex
df.groupby('Sex')['Survived'].mean()

# Multiple aggregations
df.groupby('Pclass').agg({
    'Survived': 'mean',
    'Age': 'mean',
    'Fare': 'sum'
})
```

## Creating Columns

```python
import numpy as np

# Simple calculation
df['Fare_EUR'] = df['Fare'] * 0.92

# Conditional
df['IsAdult'] = np.where(df['Age'] >= 18, 'Adult', 'Child')

# Family size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

## Handling Missing Data

```python
# Check missing
print(df.isna().sum())

# Fill with value
df['Age'] = df['Age'].fillna(df['Age'].median())

# Drop rows with missing
df_clean = df.dropna(subset=['Age', 'Embarked'])
```

# 📚 Part 3: Gapminder Dataset

New Data to Explore (45 minutes)

## What is Gapminder?

Global development data:
- 142 countries
- Years 1952-2007 (every 5 years)
- Life expectancy, population, GDP

Made famous by Hans Rosling's TED talks!

## Loading Gapminder

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv"
gap = pd.read_csv(url)

print(gap.shape)
print(gap.columns.tolist())
gap.head()
```

## Dataset Columns

| Column | Description |
|--------|-------------|
| country | Country name |
| continent | Africa, Americas, Asia, Europe, Oceania |
| year | 1952, 1957, ... 2007 |
| lifeExp | Life expectancy |
| pop | Population |
| gdpPercap | GDP per capita |

## Task 3.1: Basic Exploration

```python
# Answer:
# 1. How many rows and columns?
# 2. How many unique countries?
# 3. What years are covered?
# 4. What's the average life expectancy?
```

## Task 3.1: Solution

```python
print(f"Shape: {gap.shape}")  # (1704, 8)
print(f"Countries: {gap['country'].nunique()}")  # 142
print(f"Years: {sorted(gap['year'].unique())}")
print(f"Avg life exp: {gap['lifeExp'].mean():.1f}")  # 59.5
```

## Task 3.2: Filtering

```python
# Find:
# 1. Data for Slovenia
slovenia = gap[gap['country'] == 'Slovenia']

# 2. Data for year 2007
data_2007 = gap[gap['year'] == 2007]

# 3. European countries in 2007
europe_2007 = gap[(gap['continent'] == 'Europe') & (gap['year'] == 2007)]

# 4. Countries with life exp > 80 in 2007
long_life = data_2007[data_2007['lifeExp'] > 80]
print(long_life[['country', 'lifeExp']])
```

## Task 3.3: GroupBy Analysis

```python
# Life expectancy by continent
print(gap.groupby('continent')['lifeExp'].mean().round(1))

# GDP by continent
print(gap.groupby('continent')['gdpPercap'].mean().round(0))

# World population by year
print(gap.groupby('year')['pop'].sum())
```

## Task 3.4: Country Over Time

```python
# Slovenia's development
slovenia = gap[gap['country'] == 'Slovenia'].sort_values('year')
print(slovenia[['year', 'lifeExp', 'gdpPercap', 'pop']])

# Calculate change
first = slovenia.iloc[0]
last = slovenia.iloc[-1]
print(f"\nLife exp: {first['lifeExp']:.1f} → {last['lifeExp']:.1f}")
print(f"GDP: ${first['gdpPercap']:.0f} → ${last['gdpPercap']:.0f}")
```

# 📚 Part 4: Data Cleaning

Fixing Messy Data (45 minutes)

## Why Data Cleaning?

Real-world data problems:
- Missing values
- Duplicate rows
- Inconsistent text (case, whitespace)
- Wrong data types
- Invalid values

**80% of data science is data cleaning!**

## Example Messy Data

```python
import numpy as np

messy = pd.DataFrame({
    'Name': ['  Alice  ', 'bob', 'CHARLIE', 'Diana', 'bob'],
    'Age': ['25', '30', 'unknown', '35', '30'],
    'Salary': [50000, 60000, 55000, np.nan, 60000],
    'Dept': ['IT', 'it', 'IT', 'HR', 'IT']
})
print(messy)
```

## Problem 1: Duplicates

```python
# Check for duplicates
print(f"Duplicates: {messy.duplicated().sum()}")

# View duplicates
print(messy[messy.duplicated()])

# Remove duplicates
clean = messy.drop_duplicates()
print(f"After: {len(clean)} rows")
```

## Problem 2: Whitespace

```python
# Before
print(messy['Name'].tolist())
# ['  Alice  ', 'bob', 'CHARLIE', 'Diana', 'bob']

# Fix: strip whitespace
messy['Name'] = messy['Name'].str.strip()

# After
print(messy['Name'].tolist())
# ['Alice', 'bob', 'CHARLIE', 'Diana', 'bob']
```

## Problem 3: Inconsistent Case

```python
# Before
print(messy['Name'].tolist())  # Mixed case
print(messy['Dept'].value_counts())  # IT, it, HR

# Fix: standardize
messy['Name'] = messy['Name'].str.title()
messy['Dept'] = messy['Dept'].str.upper()

# After
print(messy['Name'].tolist())  # Alice, Bob, Charlie...
print(messy['Dept'].value_counts())  # IT, HR
```

## Problem 4: Wrong Type

```python
# Age is string!
print(messy['Age'].dtype)  # object

# Convert (errors become NaN)
messy['Age'] = pd.to_numeric(messy['Age'], errors='coerce')

print(messy['Age'].dtype)  # float64
print(messy['Age'])  # Numbers + NaN
```

## Problem 5: Missing Values

```python
# Check
print(messy.isna().sum())

# Fill with median
messy['Age'] = messy['Age'].fillna(messy['Age'].median())
messy['Salary'] = messy['Salary'].fillna(messy['Salary'].median())

# Verify
print(messy.isna().sum())  # All zeros
```

## Data Cleaning Checklist

1. ✅ `df.shape` - Check size
2. ✅ `df.dtypes` - Check types
3. ✅ `df.duplicated().sum()` - Check duplicates
4. ✅ `df.isna().sum()` - Check missing
5. ✅ `.str.strip()` - Remove whitespace
6. ✅ `.str.lower()` or `.str.title()` - Standardize case
7. ✅ `pd.to_numeric()` - Convert types
8. ✅ `.fillna()` - Handle missing

## Exercise 4.1: Clean This Data

```python
dirty = pd.DataFrame({
    'product': ['  Apple', 'BANANA', 'apple', 'Cherry  '],
    'price': ['1.50', '0.75', '1.50', 'N/A'],
    'quantity': [10, 15, 10, 8]
})

# Tasks:
# 1. Strip whitespace
# 2. Standardize to title case
# 3. Convert price to numeric
# 4. Remove duplicates
```

# 📚 Part 5: Practice Challenges

Test Your Skills (30 minutes)

## Challenge 1: Top Countries

Using Gapminder 2007:
- Find top 5 countries by life expectancy
- Find bottom 5 countries
- Calculate the difference

## Challenge 1: Solution

```python
gap_2007 = gap[gap['year'] == 2007]

# Top 5
top_5 = gap_2007.nlargest(5, 'lifeExp')[['country', 'lifeExp']]
print("Top 5:")
print(top_5)

# Bottom 5
bottom_5 = gap_2007.nsmallest(5, 'lifeExp')[['country', 'lifeExp']]
print("\nBottom 5:")
print(bottom_5)

# Difference
diff = gap_2007['lifeExp'].max() - gap_2007['lifeExp'].min()
print(f"\nDifference: {diff:.1f} years")
```

## Challenge 2: Continent Summary

For 2007, create a summary by continent:
- Number of countries
- Average life expectancy
- Average GDP
- Total population

## Challenge 2: Solution

```python
gap_2007 = gap[gap['year'] == 2007]

summary = gap_2007.groupby('continent').agg({
    'country': 'count',
    'lifeExp': 'mean',
    'gdpPercap': 'mean',
    'pop': 'sum'
}).round(1)

summary.columns = ['Countries', 'Life Exp', 'GDP', 'Population']
print(summary)
```

## Challenge 3: Growth Analysis

For European countries:
- Calculate life exp change from 1952 to 2007
- Find most improved country
- Find least improved country

## Challenge 3: Solution

```python
europe = gap[gap['continent'] == 'Europe']

# Get 1952 and 2007
e1952 = europe[europe['year'] == 1952][['country', 'lifeExp']]
e2007 = europe[europe['year'] == 2007][['country', 'lifeExp']]

# Merge and calculate
merged = e1952.merge(e2007, on='country', suffixes=('_1952', '_2007'))
merged['change'] = merged['lifeExp_2007'] - merged['lifeExp_1952']

print("Most improved:")
print(merged.nlargest(3, 'change')[['country', 'change']])

print("\nLeast improved:")
print(merged.nsmallest(3, 'change')[['country', 'change']])
```
